## **EDA**

Exploratory Data Analysis (EDA) is an important first step in data science projects. It involves looking at and visualizing data to understand its main features, find patterns, and discover how different parts of the data are connected.

```python
country_names=final_df.Country.value_counts().index
```

By this we can say that the zomato has more order in india then USA then UK.

```python
country_values=final_df.Country.value_counts().values
```

```python
## Pie chart
plt.pie(country_values[:3], labels=country_names[:3])
plt.show()
```

Observation major transcation in of Zomato is occur in Indian >>> USA >>> UK

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0: 'Rating Count'})
```

### **Observation**

0 --> White --> not rated
1.8 - 2.4 ---> Red --> Poor
2.5 - 3.4 --> Orange --> Avg
3.5 - 3.9 --> Yellow --> Good
4.0 - 4.4 --> Green ---> Very Gud
4.5 - 4.9 --> Dark Green --> Excellent

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings)
plt.show()
```

From my obeservation the count of people are giving the rating to any resturant which has the rating 2.5 - 3.4

and also not giving rating count is highest on the less rated hotel
